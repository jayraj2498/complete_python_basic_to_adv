```python
class Library:
    def __init__(self, listbooks):
        self.books = listbooks

    def available_books(self):
        print("Available books in our stocks are:  ")

        for book in self.books:
            print("  *" + book)

    def borrow_book(self, bookname):
        if bookname in self.books:
            print(f"You have been issued {bookname}. Please keep it safe and return it in 30 days.")
            self.books.remove(bookname)
            return True
        else:
            print(f"Sorry, this {bookname}book is not available because it not valid book \n Or.. issued to someone else.")
            return False

    def return_book(self, bookname):
        self.books.append(bookname)
        print(f"Thank you for returning the {bookname} book.")


class Student:
    def request_book(self):
        book = input("Enter the book name that you want to borrow from Liabrary : ")
        return book

    def return_book(self):
        book = input("Enter the book name that you want to return to Liabrary : ")
        return book


if __name__ == "__main__":
    city_centre_library = Library(["algebra", "geometry", "english", "science", "history"])
    student = Student()

    while True:
        welcome_msg = '''\n =====Welcome to Library System=====
        1. List Of All Books
        2. Request Books
        3. Return Books
        4. Exit Library \n 
        '''
        
        print(welcome_msg )
        
        choice = int(input("Enter your choice: "))

        if choice == 1:
            city_centre_library.available_books()
        elif choice == 2:
            city_centre_library.borrow_book(student.request_book())
        elif choice == 3:
            city_centre_library.return_book(student.return_book())
        elif choice == 4:
            print("Thanks for visiting our Library.")
            break
        else:
            print("Please Enter the Valid choice your choice is Invalid .")
```

```python
# Student Database management system  project 


class StudentDatabase:
    def __init__(self):
        self.students = {}

    def add_student(self, student_id, name, age, grade):
        """
        Add a new student to the database.

        Parameters:
        - student_id: Unique identifier for the student.
        - name: Name of the student.
        - age: Age of the student.
        - grade: Grade of the student.
        """
        if student_id not in self.students:
            self.students[student_id] = {'name': name, 'age': age, 'grade': grade}
            print(f"Student {name} added to the database.")
        else:
            print(f"Student with ID {student_id} already exists. Use edit_student to update information.")

    def edit_student(self, student_id, name, age, grade):
        """
        Edit existing student information.

        Parameters:
        - student_id: Unique identifier for the student.
        - name: Updated name of the student.
        - age: Updated age of the student.
        - grade: Updated grade of the student.
        """
        if student_id in self.students:
            self.students[student_id] = {'name': name, 'age': age, 'grade': grade}
            print(f"Student {name}'s information updated.")
        else:
            print(f"Student with ID {student_id} not found. Use add_student to add a new student.")

    def display_students(self):
        """Display information for all students in the database."""
        if not self.students:
            print("No students in the database.")
        else:
            print("Student Database:")
            for student_id, info in self.students.items():
                print(f"ID: {student_id}, Name: {info['name']}, Age: {info['age']}, Grade: {info['grade']}")

# Create an instance of the StudentDatabase
student_db = StudentDatabase()

while True:
    print("\nOptions:")
    print("1. Add Student")
    print("2. Edit Student")
    print("3. Display Students")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        student_id = input("Enter student ID: ")
        name = input("Enter student name: ")
        age = input("Enter student age: ")
        grade = input("Enter student grade: ")
        student_db.add_student(student_id, name, age, grade)

    elif choice == '2':
        student_id = input("Enter student ID to edit: ")
        name = input("Enter updated name: ")
        age = input("Enter updated age: ")
        grade = input("Enter updated grade: ")
        student_db.edit_student(student_id, name, age, grade)

    elif choice == '3':
        student_db.display_students()

    elif choice == '4':
        print("Exiting the student database. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 4.") 
        
        
        
        
'''
    
Explanation:

The StudentDatabase class is created to manage the student information.
The add_student method adds a new student to the database.
The edit_student method updates the information for an existing student.
The display_students method displays information for all students in the database.
In the main part of the code, a loop allows the user to choose options (add student, edit student, display students, or exit).
User input is used to perform the desired actions based on the selected option.
'''
```

```python
class SimpleCalculator:
    def add(self, num1, num2):
        """Perform addition."""
        return num1 + num2

    def subtract(self, num1, num2):
        """Perform subtraction."""
        return num1 - num2

    def multiply(self, num1, num2):
        """Perform multiplication."""
        return num1 * num2

    def divide(self, num1, num2):
        """
        Perform division.
        If the divisor is 0, display an error message and return None.
        """
        if num2 != 0:
            return num1 / num2
        else:
            print("Error: Division by zero.")
            return None

# Create an instance of the SimpleCalculator
calculator = SimpleCalculator()

while True:
    print("\nOptions:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        if choice == '1':
            result = calculator.add(num1, num2)
            print(f"Result: {num1} + {num2} = {result}")

        elif choice == '2':
            result = calculator.subtract(num1, num2)
            print(f"Result: {num1} - {num2} = {result}")

        elif choice == '3':
            result = calculator.multiply(num1, num2)
            print(f"Result: {num1} * {num2} = {result}")

        elif choice == '4':
            result = calculator.divide(num1, num2)
            if result is not None:
                print(f"Result: {num1} / {num2} = {result}")

    elif choice == '5':
        print("Exiting the calculator. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
